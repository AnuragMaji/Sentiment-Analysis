```python
import csv
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import LinearSVC
from sklearn.multiclass import OneVsRestClassifier, OneVsOneClassifier
import numpy as np
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
from nltk.stem import SnowballStemmer
```

```python
train_set = []
test_set = []
```

```python
with open('train.tsv', 'r') as f:
    sentimentreader = csv.reader(f, delimiter='\t')
    header = next(sentimentreader)
    cnt = 0
    for row in sentimentreader:
        sentence = row[2]
        sentiment = row[3]
        if (cnt < 30):
            test_set.append((sentence, sentiment))
        elif (cnt > 30):
            train_set.append((sentence, sentiment))
        cnt += 1
```

```python
train_set
```

```python
sentence_set = [sentence for sentence, label in train_set]
label_set = [int(label) for sentence, label in train_set]
```

```python
from sklearn import model_selection
Xtrain,Xtest,Ytrain,Ytest = model_selection.train_test_split(sentence_set,label_set,test_size=0.3,random_state=1701)
```

```python
# print(len(Xtrain))
# print(len(Ytrain))
# print(len(Xtest))
# print(len(Ytest))
```

# Support Vector Classifier

```python
text_clf_svm = Pipeline([('vect', CountVectorizer()),
                     ('tfidf', TfidfTransformer()),
                     ('clf', LinearSVC()),
                     ])
X_train = np.asarray(Xtrain)
text_class_svm = text_clf_svm.fit(X_train, np.asarray(Ytrain))
print(text_class_svm)
```

```python
pred_train_svm = text_class_svm.predict(np.asarray(Xtrain))
# Test to check whether our classifier makes prediction across all categories
print(np.unique(pred_train_svm))  
pred_test_svm = text_class_svm.predict(np.asarray(Xtest))
# Test to check whether our classifier makes prediction across all categories
print(np.unique(pred_test_svm)) 
```

```python
from sklearn.metrics import classification_report
target_names = ['negative-0', 'somewhat negative-1', 'neutral-2','somewhat positive-3','positive-4']
print(classification_report(y_pred = pred_train_svm, y_true=np.asarray(Ytrain), target_names=target_names))
print(classification_report(y_pred = pred_test_svm, y_true=np.asarray(Ytest), target_names=target_names))
```
